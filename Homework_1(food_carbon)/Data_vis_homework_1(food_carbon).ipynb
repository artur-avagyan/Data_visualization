### Import all the packeges we need for our homework

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_palette("gnuplot_r")
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

### Read csv 

```python
data_food=pd.read_csv('food_consumption.csv')
data_food
```

```python
data_food[data_food['country'].isin(['Armenia','Bangladesh'])]
```

### Show categories from 'food_category' column and print length of that

```python
print(data_food['food_category'].value_counts().index)
print('Length is --> ',len(data_food['food_category'].value_counts().index))
```

### Show distribution of consumptions from each category

```python
fig, axes = plt.subplots(6, 2, figsize=(20,35))
fig.delaxes(axes[5,1])
fig.suptitle('Distributions of consumptions', fontsize=50)
each_type=data_food['food_category'].value_counts().index
numb=0
for i in range(6):
    for j in range(2):
        if i==5 and j==1:
            continue
        data_sup=data_food[data_food['food_category']==each_type[numb]]['consumption']
        sns.distplot(ax=axes[i,j],a=data_sup)
        axes[i,j].set_title(each_type[numb])
        numb+=1
plt.savefig('Distributions of consumptions.png')
```

### Show distribution of CO2 in consumptions from each category

```python
fig, axes = plt.subplots(6, 2, figsize=(20,35))
fig.delaxes(axes[5,1])
fig.suptitle('Distributions of CO2 emmission', fontsize=50)
each_type=data_food['food_category'].value_counts().index
numb=0
for i in range(6):
    for j in range(2):
        if i==5 and j==1:
            continue
        data_sup=data_food[data_food['food_category']==each_type[numb]]['co2_emmission']
        sns.distplot(ax=axes[i,j],a=data_sup)
        axes[i,j].set_title(each_type[numb])
        numb+=1

plt.savefig('Distributions of CO2 emmission.png')
```

### Show barplot of Armenia's comsuption

```python
plt.figure(figsize=(20,10))
data_arm=data_food[data_food['country']=='Armenia']
sns.barplot(data=data_arm,x='food_category',y='consumption',palette='gnuplot_r')
plt.show()
```

### Show top 10 countries with most fish consumption value

```python
plt.figure(figsize=(25,15))
data_fish=data_food[data_food['food_category']=='Fish'].sort_values(by=['consumption']).tail(10)
sns.barplot(data=data_fish,y='country',x='consumption',palette='gnuplot_r',
            order=data_fish.sort_values('consumption',ascending = False).country)
for index, value in enumerate(sorted(data_fish['consumption'],reverse=True)): #show values in figure
    plt.text(value, index, str(value))
plt.show()
```
