### Information about data

This is a set of just over 20,000 games collected from a selection of users on the site Lichess.org, and how to collect more. This set contains the:

- **id**             ---> Game ID
- **rated**          ---> Rated (T/F)
- **created_at**     ---> Start Time
- **last_move_at**   ---> End Time
- **turns**          ---> Number of Turns
- **victory_status** ---> Game Status
- **winner**         ---> Winner
- **increment_code** ---> Time Increment
- **white_id**       ---> White Player ID
- **white_rating**   ---> White Player Rating
- **black_id**       ---> Black Player ID
- **black_rating**   ---> Black Player Rating
- **moves**          ---> All Moves in Standard Chess Notation
- **opening_eco**    ---> Opening Eco (Standardised Code for any given opening, list here)
- **opening_name**   ---> Opening Name
- **opening_ply**    ---> Opening Ply (Number of moves in the opening phase)

### Create column, that will present who has higher rating

```python
conditions = [
    games_df['white_rating'] > games_df['black_rating'],
    games_df['white_rating'] < games_df['black_rating'],
    games_df['white_rating'] == games_df['black_rating']
    ]

# create a list of the values we want to assign for each condition
values = ['white', 'black', 'draw']

# create a new column and use np.select to assign values to it using our lists as arguments
games_df['high_rating'] = np.select(conditions, values)
```

```python
import random
cross_df=pd.crosstab(games_df['winner'],games_df['high_rating'])
plt.figure(figsize=(10,7))
sns.heatmap(cross_df/np.sum(cross_df,axis=0), annot=True,fmt='.2%', cmap=random.choice(colors_list))
plt.title("Crosstab of High rating and Winner")
plt.show()
```

According to this table if `black` has a high rating, they will have $60.17$% probability of winning. In the case of high rating belong to `white`, then they have a $64.12$% probablity of winning. In case of equal ratings, the probability of winning the `black` and `white` is equal: $47$%

### For 3rd hypotes group by data by `opening_name` column, then barplot first 10 minimum turns

```python
open_mean_turns=games_df.groupby('opening_name')['turns'].agg(mean_turns='mean').reset_index()
head_data=open_mean_turns.sort_values(by=['mean_turns']).head(10)
plt.figure(figsize=(11,7))
sns.barplot(x="mean_turns", 
    y="opening_name",
    data=head_data,palette=random.choice(colors_list),
    ci=None)
for index, value in enumerate(head_data['mean_turns']): #show values in figure
    plt.text(value, index, str(value))
plt.title("Mean turns of each Opening name\n(Tail 10)",size=15)
plt.ylabel("")
plt.xlabel("Mean of each type",size=15)
plt.show()
```

### Analys this by ratings

```python
group_data_opening=games_df.groupby('opening_name').agg(mean_turns=('turns', "mean"),
                                                        mean_white_rating=('white_rating', "mean"),
                                                        mean_black_rating=('black_rating', "mean"),
                                                        most_winner=("winner", pd.Series.mode)    
).reset_index().sort_values(by=['mean_turns'])
```

```python
fig, axes = plt.subplots(1, 2, figsize=(20,10))
fig.suptitle('Turns vs Ratings', fontsize=40,size=20)
sns.scatterplot(x='mean_white_rating',y='mean_turns',data=group_data_opening,ax=axes[0])
sns.scatterplot(x='mean_black_rating',y='mean_turns',data=group_data_opening,ax=axes[1])
axes[0].set_title("Mean Turns vs Mean White Ratings",size=15)
axes[1].set_title("Turns vs Black Ratings",size=15)
axes[0].set_ylabel("Mean White Ratings",size=15)
axes[0].set_xlabel("Mean White Rating",size=15)
axes[1].set_ylabel("Mean Black Ratings",size=15)
axes[1].set_xlabel("Mean Black Rating",size=15)
plt.show()
```

There are no difference by ratings
